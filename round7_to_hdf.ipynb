```python
import pandas as pd

%time df = pd.read_spss('r7_merged_data_34ctry.release.sav')
```

```python
for name in df.columns:
    print(name)
```

```python
del df['STRTIME']
```

```python
del df['ENDTIME']
```

```python
del df['LENGTH']
```

```python
df['DATEINTR'] = pd.to_datetime(df['DATEINTR'])
```

```python
df['Q1'] = pd.to_numeric(df['Q1'], errors='coerce')
```

```python
for i, column in enumerate(df.columns):
    name = 'column%.3d' % i
    try:
        df[column].to_hdf('test.hdf5', name, format='table')
    except:
        print(name, column)        
```
